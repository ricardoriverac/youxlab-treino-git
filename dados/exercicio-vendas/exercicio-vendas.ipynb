```python
import pandas as pd
df = pd.read_csv('/home/youx/projeto-git/youxlab-treino-git/dados/exercicio-vendas/exercicio_vendas.csv')
df.dropna(inplace=True)
df.head()
del df['Unnamed: 0']
df
```

```python
grupos = df.groupby(['Região', 'Produtos']).mean().reset_index()
grupos
```

```python
def nota(vendas):
    if vendas < 3000:
        return 'Baixa'
    elif 3000 <= vendas <= 7000:
        return 'Média'
    else:
        return 'Alta' 
      
df['Faixa de Vendas'] = df['Vendas'].apply(nota)
df
```

```python
produtos_regiao = df.groupby(['Região', 'Faixa de Vendas'])['Produtos'].value_counts()
produtos_regiao
```

```python
baixa = df.query('`Faixa de Vendas` == "Baixa"').reset_index()
baixa
```

```python
medio = df.query('`Faixa de Vendas` == "Média"').reset_index()
medio
```

```python
alta = df.query('`Faixa de Vendas` == "Alta"')
alta
```
