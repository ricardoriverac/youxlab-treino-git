FAZENDO A DATA


```python
import pandas as pd
data = pd.read_csv('./datasets/GasPricesinBrazil_2004-2019.csv', sep=";")
data
```

```python
data.head()
```

ATRIBUINDO DADOS

```python
# atribuindo constantes
# retorna uma referência/view da coluna produto 
produto_view = data['PRODUTO']
```

```python
produto_copy = data['PRODUTO'].copy() # retorna uma cópia da coluna 'PRODUTO'
produto_copy
```

```python
data['PRODUTO'] = 'Combustível' # atribuindo o valor 'Combustível' para linha do dataframe na coluna 'PRODUTO'
```

```python
nrows, ncols = data.shape
```

```python
# atribuindo listas ou series

novos_produtos = [f'Produto {i}' for i in range(nrows)]
len(novos_produtos)
```

```python
# a quantidade de elementos da lista 'novos_produtos' é igual ao número de linhas do dataframe
data['PRODUTO'] = novos_produtos
```

```python
print(produto_view)
print('\n')
print(produto_copy)
```

```python
# voltando para os produtos originais
data['PRODUTO'] = produto_copy
data
```

CRIANDO NOVAS COLUNAS


```python
# criando uma coluna a partir de um valor constatnte/default
# todas as linhas terão o mesmo valor 'DEFAULT'
data['coluna sem nocao'] = 'DEFAULT'
data
```

```python
data['coluna a partir de lista'] = range(data.shape[0])
data
```

```python
# nao funciona pois a quantidade de elementos da lista (a serem atribuídos a nova coluna) é diferente da qtd de linhas do dataframe
data['nao funciona'] =  [1,2,3]
```

```python
data.head()
```

```python
data['PREÇO MÉDIO REVENDA (dólares)'] = data['PREÇO MÉDIO REVENDA'] * 6.0
data
```
